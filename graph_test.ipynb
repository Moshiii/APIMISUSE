```python
import matplotlib.pyplot as plt
import numpy as np

operation = (
    "Missing",
    "Redundant",
    "Replace",
    "Outdated",
)
elements = {
    "API Method":    np.array([114,120,123,90]),
    "API Parameter": np.array([87,58,50,123]),
    "API Condition": np.array([40,28,16,35]),
}
width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(4)
bottom_2 = np.zeros(4)

for boolean, element_count in elements.items():
    p = ax.bar(operation, element_count, width, label=boolean, bottom=bottom)
    bottom += element_count
    ax.bar_label(p, label_type='center')  # Add number labels to each stack box

ax.set_ylim(0, 350)
ax.set_title("Number of API Misuse Elements per API Operation")
ax.set_ylabel("Count")
ax.set_xlabel("Operation")
ax.legend(loc="upper left")
# save to pdf
plt.savefig("API_Misuse_Elements_per_API_Operation.pdf", bbox_inches='tight')


plt.show()
```
